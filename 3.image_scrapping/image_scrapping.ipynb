```python
import bs4
```

```python
import requests
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
r = requests.get(url)
```

```python
print(r.content)
```

```python
soup = bs4.BeautifulSoup(r.content)
```

```python
print(soup)
```

```python
img_element = soup.findAll('img')
img_element = img_element[2:]
```

```python
img_element[32]
```

```python
img = img_element[2]
```

```python
img.attrs['src']
```

```python
with open("inspiration.jpg","wb") as file:
    img_url = img.attrs['src']
    r = requests.get(img_url)
    file.write(r.content)
```

```python
for i,img in enumerate(img_element):
    with open("inspiration{}.jpg".format(i),"wb") as file:
         img_url = img.attrs['src']
         r = requests.get(img_url)
         file.write(r.content)
```
